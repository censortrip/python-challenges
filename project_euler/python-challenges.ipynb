## Problem 1: Multiples of 3 and 5
---
If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9.

The sum of these multiples is 23. Find the sum of all the multiples of 3 or 5 below 1000.

```python
#My solution:
s = 0

for x in range(1,1000):
    if x%3==0 or x%5==0:
        s += x

print(s)

# More succinct solution
# list_ = [i for i in range(3,1000) if i%3==0 or i%5==0]
# print(sum(list_))
```

## Problem 2: Even Fibonacci numbers
---

## Problem 6: Sum square difference
---
The sum of the squares of the first ten natural numbers is:

$ 1^2 + 2^2 + ... + 10^2 = 385 $

The square of the sum of the first ten natural numbers is:

$ (1 + 2+ ... + 10)^2 = 55^2 = 3025 $

Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is:

$ 3025 − 385 = 2640 $ 

Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

```python
squares = 0
sumz = 0

t0 = time.time()

for i in range(1, 101):
    squares += i**2
    sumz += i

print(sumz**2 - squares)

t1 = time.time()
print("Executed in: " + str(t1-t0) + " Seconds")
```

## Problem 7: 10001st prime
---
By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.

What is the 10,001st prime number?


```python
count = 6
number = 0

# Note a variable state is created with initial value True. 
# This is needed so that if a number is divisible by another other than 1
# state can be set to False. 
# This allows a conditional statement to test the function for true or false
def isPrime(num):
    state = True
    for i in range(3, int(math.sqrt(num))+1):
        if num%i==0:
            state = False
            break
    return(state)
    
t0 = time.time()

for i in range(15, 1000000, 2):
    if isPrime(i) == True and count < 10001:
        count += 1
        number = i

print("Final solution at point count: " + str(count) + " is: " + str(number))
t1 = time.time()
print("Executed in: " + str(t1-t0) + " Seconds")
```

```python
# Solution 2
# Found on stack overflow: https://stackoverflow.com/questions/3939660/sieve-of-eratosthenes-finding-primes-python
# A much more efficient method than above using the Seive of Eratosthenes
# Algorithm for locating all primes below of a given number

def eratosthenes(n):
    non_primes =  set()  
    primes = set()
    
    # Start loop at 2 as 0, and 1 are not primes
    for i in range(2, n+1):
        
        # If i is not in non_primes, then it is a prime and needs to be added
        # to primes set
        if i not in non_primes:
            primes.add(i)
            
            # This section loops through all the multiples of i and adds
            # them to not_primes, so when this finishes and i loop restarts
            # if i is not in prime list, then it's prime.
            for j in range(i*i, n+1, i):
                non_primes.add(j)    # add() is used to append to a set
    #print(primes)
    print(sum(primes))
    

t0 = time.time()

num = 2000000  
eratosthenes(num)

t1 = time.time()
print("Executed in: " + str(t1-t0) + " Seconds")
```

```python
# Solution 3
# Found in solutions for the problme on Project Euler

# An extremely efficient solution using the Sieve of Eratosthenes is 
# a very clever manner.

# This uses the index as the numbers to check, not the initial set up
# creates a list of 2 million True Statements and sets index 0 and 1 as false
# as 0 and 1 are not primes
def sieve(n):
    primes=[True]*(n+1)
    primes[0]=False
    primes[1]=False
    res=[]
    
    for i, prime in enumerate(primes):
        if prime: 
            primes[i**2:n+1:i]=[False]*len(primes[i**2:n+1:i])
            res.append(i)
    return sum(res)


t0 = time.time()

print(sieve(2000000))

t1 = time.time()
print("Executed in: " + str(t1-t0) + " Seconds")
```
