## Problem 6: Sum square difference

The sum of the squares of the first ten natural numbers is:

$ 1^2 + 2^2 + ... + 10^2 = 385 $

The square of the sum of the first ten natural numbers is:

$ (1 + 2+ ... + 10)^2 = 55^2 = 3025 $

Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is:

$ 3025 − 385 = 2640 $ 

Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

```python
# Solution 2
# Found on stack overflow: https://stackoverflow.com/questions/3939660/sieve-of-eratosthenes-finding-primes-python
# A much more efficient method than above using the Seive of Eratosthenes
# Algorithm for locating all primes below of a given number

def eratosthenes(n):
    non_primes =  set()  
    primes = set()
    
    # Start loop at 2 as 0, and 1 are not primes
    for i in range(2, n+1):
        
        # If i is not in non_primes, then it is a prime and needs to be added
        # to primes set
        if i not in non_primes:
            primes.add(i)
            
            # This section loops through all the multiples of i and adds
            # them to not_primes, so when this finishes and i loop restarts
            # if i is not in prime list, then it's prime.
            for j in range(i*i, n+1, i):
                non_primes.add(j)    # add() is used to append to a set
    #print(primes)
    print(sum(primes))
    

t0 = time.time()

num = 2000000  
eratosthenes(num)

t1 = time.time()
print("Executed in: " + str(t1-t0) + " Seconds")
```

```python
# Solution 3
# Found in solutions for the problme on Project Euler

# An extremely efficient solution using the Sieve of Eratosthenes is 
# a very clever manner.

# This uses the index as the numbers to check, not the initial set up
# creates a list of 2 million True Statements and sets index 0 and 1 as false
# as 0 and 1 are not primes
def sieve(n):
    primes=[True]*(n+1)
    primes[0]=False
    primes[1]=False
    res=[]
    
    for i, prime in enumerate(primes):
        if prime: 
            primes[i**2:n+1:i]=[False]*len(primes[i**2:n+1:i])
            res.append(i)
    return sum(res)


t0 = time.time()

print(sieve(2000000))

t1 = time.time()
print("Executed in: " + str(t1-t0) + " Seconds")
```

## Problem 14: Longest Collatz sequence

The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)
n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:
13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1

It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.


```python
import time

def collatz(num):
    seq = [num]  
    while num > 1:
        if num % 2 == 0 :
            num //= 2
            seq += [num]
        else: 
            num = (3 * num + 1) // 2
            seq += [num]     
    return(len(seq))         
             
start = time.time()
   
cur_max = 0
solution = 0
for i in range(1, 1_000_000, 2):
    seq_len = collatz(i)
    if seq_len > cur_max:
        cur_max = seq_len
        solution = i

print(f'Solution: {solution}')        
print(time.time() - start)
```

```python
import time

cache = {}

def collatz(num):
    seq_len = 0
    while num > 1:
        if num in cache:
            seq_len += cache[num]
            break
        else:
            if num % 2 == 0:
                num = num // 2
            else:
                num = num * 3 + 1
            seq_len += 1
    return(seq_len)   


def prob15(num):
    cur_max = 0
    solution = 0
    for i in range(1, num+1):
        cache[i] = collatz(i)   # cache, keys = i, values = seq_length
        if cache[i] > cur_max:
            cur_max = cache[i]
            solution = i
    

start = time.time()

prob15(1_000_000)

print(f'Solution: {solution}') 
print(f'Solution Sequence Length: {cache[837799]}')
print(f'Finished In: {time.time() - start} sec')
```

## Problem 15: Lattice paths

Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.

How many such routes are there through a 20×20 grid?


Solution: 

Each move is either right or down and so can be represented as 1 or 0. Since the move is from one corner of a 20x20 grid to the opposite corner a total of 40 moves is required. 

A path, then, is logically equivalent to a 40 bit number 20 of whose bits are 0 and 20 are 1. So the number of paths is the number of permutations of 20 zeroes and 20 ones. 

That formula is taken from the binomial theorem, 40!/(20! * 20!). There are any number of ways to evaluate this number, some easier than others.

```python
import math

n = 20

#solution = math.factorial(40)/(math.factorial(20) * math.factorial(20))
solution = math.factorial(n*2) / (math.factorial(n) **2)

print(solution)
```

## Problem 17: Number letter counts

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?

NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.


```python
def get1digit(num):
    one_d = {1 : 'one', 2 : 'two', 3 : 'three', 4 : 'four', 5 : 'five',
             6 : 'six', 7 : 'seven', 8 : 'eight', 9 : 'nine'}
    return(one_d[num])

def get2digit(num):
    two_d = {10 : 'ten', 11 : 'eleven', 12 : 'twelve', 13 : 'thirteen',
             14 : 'fourteen', 15 : 'fifteen', 16 : 'sixteen', 17 : 'seventeen',
             18 : 'eighteen', 19 : 'nineteen', 20  : 'twenty', 30  : 'thirty',
             40  : 'forty', 50  : 'fifty', 60  : 'sixty', 70  : 'seventy',
             80  : 'eighty', 90  : 'ninety'}
    
    if (len(str(num)) == 2 and num < 20) | (len(str(num)) == 2 and num % 10 == 0):
        return(two_d[num])
    else:
        return (f'{two_d[math.floor(num/10)*10]}{get1digit(int(str(num)[1]))}')
    
def get3digit(num):
    if num % 100 == 0:
        return (f'{get1digit(math.floor(num/100))}hundred')
    elif math.floor(num%100) < 10:
        return (f'{get1digit(math.floor(num/100))}hundredand{get1digit(math.floor(num%100))}')
    else:
        return (f'{get1digit(math.floor(num/100))}hundredand{get2digit(math.floor(num%100))}')
    
def prob17(max_val):
    answer = 0
    for i in range(1, max_val):
        if len(str(i)) == 1:
            answer += len(get1digit(i))    
        elif len(str(i)) == 2:
            answer += len(get2digit(i))
        elif len(str(i)) == 3:
            answer += len(get3digit(i))    
        else:
            s = 'onethousand'
            answer += len(s)
            
    return answer

s = time.time()
prob17(1001)  
print(time.time() - s)
```

## Problem 19: Counting Sundays


You are given the following information, but you may prefer to do some research for yourself.

* 1 Jan 1900 was a Monday.
* 30 day monhts are: September, April, June and November
* 31 day months are all the rest aside from February
* February has 28 days on normal years and 29 days on leap years
* A leap year occurs on any year evenly divisible by 4, but not on a century unless it is divisible by 400.

How many Sundays fell on the first of the month during the twentieth century (1 Jan 1901 to 31 Dec 2000)?


```python
import datetime as dt

counter = 0
year = 1901
month = 1

day = dt.date(year, month, 1)

while day.year < 2001:
    if day.weekday() == 6:
        counter += 1
        
    if month+1 == 13:
        month = 1
        year += 1
    else:
        month += 1
        
    day = dt.date(year, month, 1)
        
print(counter)  
```

## Problem 20: Factorial digit sum

n! means n × (n − 1) × ... × 3 × 2 × 1

For example, 10! = 10 × 9 × ... × 3 × 2 × 1 = 3628800,
and the sum of the digits in the number 10! is 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27.

Find the sum of the digits in the number 100!


```python
solution = 0 
number = math.factorial(100)

for num in str(number):
    solution += int(num)
    
print(f'\nSolution: {solution}')
```
